```python
#Import teh libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create a data set inorder to apply linear regression.
X = np.random.uniform(-4,4,500)
Y = X + np.random.standard_normal(500)+2.5
plt.plot(X, Y, 'o')
plt.show()
```

```python
def linear_regression(X, y, w, b, learning_rate = 0.01, epochs = 1000):
    m = float(len(Y))
    cost_history = {}
    for i in range(epochs):
        #Calculate the prediction
        y_hat = np.dot(w, X) + b
        #Calculate the cost
        cost = (1/(2*m)) * np.sum((y_hat - y) ** 2)
        cost_history[i] = cost
        
        #Compute the gradients
        w_gradient = (1/m) * np.sum(X * (y_hat - y))
        b_gradient = (1/m) * np.sum(y_hat - y)
        #Update the weight and bias using Gradient Descent
        w = w - (learning_rate * w_gradient)
        b = b - (learning_rate * b_gradient)
    return w, b, cost_history   
    
```

```python
w, b, cost_history = linear_regression(X, Y, 0, 0)
#Plot the cost function
for i in cost_history:
    plt.plot(i, cost_history[i], 'o')
plt.show()    
```
