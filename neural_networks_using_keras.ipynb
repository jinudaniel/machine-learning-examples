```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
seed = 7
np.random.seed(seed)
```

```python
# load pima indians dataset
dataset = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data", 
                      header=None).values
```

```python
dataset[0:5,:]
```

```python
# Split the dataset into train and test set
X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8], 
                                                    test_size=0.25, random_state=87)
```

```python
# form a neural network.
model = Sequential()
model.add(Dense(12,input_dim = 8, activation='relu')) # First hidden layer
model.add(Dense(8, activation='relu')) # second hidden layer
model.add(Dense(1, activation='sigmoid')) # output layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=200, batch_size=16)
```

```python
model.evaluate(X_test, Y_test)
```
